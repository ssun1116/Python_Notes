# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 출력
```

```python
df[df['키'] >= 185]
```

```python
df[df['키'] < 185]
```

```python
df[~(df['키'] >= 185)] # 조건을 역으로 적용.
```

```python
df[df['키'] >= 185]['수학']
```

```python
df.loc[df['키'] >= 185, '수학'] # loc를 이용한 필터링
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']]
```

## 다양한 조건

```python
df.loc[(df['키'] >= 185) & (df['학교'] == "북산고")]
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송'씨 성을 가진 사람
```

```python
df[filt]
```

```python
filt = df['이름'].str.contains('태') # '태'가 들어가는 사람
```

```python
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가지 않는 사람
```

```python
langs = ["Python", "Java"]
```

```python
df[df['SW특기'].isin(langs)]
```

```python
df # 황태산 -> 포함이 안됨 (대문자여서)
```

```python
df['SW특기'] = df['SW특기'].str.lower()
```

```python
df
```

```python
df[df['SW특기'].str.contains('java')] # error - Nan 데이터
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java') # 4, 5번은 NAN으로 나옴 -> 필터링 불가
```

```python
df[df['SW특기'].str.contains('java', na=True)] # nan 데이터는 모두 true로 인지한다는 뜻
```

```python
df[df['SW특기'].str.contains('java', na=False)] # nan 데이터 걸러짐
```

```python
filt = df['SW특기'].str.contains('java', na = False)
```
