# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 여부를 출력
```

```python
df[df['키'] >= 185]
```

```python
df[df['키'] < 185]
```

```python
df[~(df['키'] >= 185)] # 조건을 역으로 적용.
```

```python
df[df['키'] >= 185]['수학']
```

```python
df.loc[df['키'] >= 185, '수학'] # loc를 이용한 필터링
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']]
```

## 다양한 조건

```python
df.loc[(df['키'] >= 185) & (df['학교'] == "북산고")]
```
