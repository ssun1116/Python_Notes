## 데이터 제외하기 - dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(axis = 'index', how = 'any')
```

- axis : index (row) or columns
- how : any (하나라도 NaN이 있으면 지움) or all (row 전체가 NaN일때 지움)

```python
df.dropna(axis = 'index', how = 'any') # NaN이 하나라도 있는 row를 지움
```

```python
df.dropna(axis = 'columns', how = 'any') # 'SW특기' column이 사라짐
```
